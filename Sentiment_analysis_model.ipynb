# Data Preprocessing

```python
stopwords=stopwords.words('english')
print(stopwords)
```

```python
ps=nltk.stem.PorterStemmer()
def pre_process(text):
    words=[i.lower() for i in text.split() if i.isalpha()]          #nltk.word_tokenize(text)
    words=[ps.stem(i) for i in words if i not in stopwords]
    return ' '.join(words)
```

```python
df['tweets']=df['tweets'].apply(lambda x:pre_process(x))
print(df.head())
```

```python
def labels_encode(l):
    if l=='good':return 1
    if l=='bad':return -1
    if l=='neutral':return 0
df['labels']=df['labels'].apply(lambda x:labels_encode(x))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

from sklearn.model_selection import train_test_split

from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
X1=CountVectorizer().fit_transform(df['tweets'])
X1,y1=X1.toarray(),df['labels']
```

```python
X2=TfidfVectorizer().fit_transform(df['tweets'])
X2,y2=X2.toarray(),df['labels']
```

```python
X1_train, X1_test, Y1_train, Y1_test = train_test_split(X1, y1, test_size = 0.20, random_state = 0)
```

```python
X2_train, X2_test, Y2_train, Y2_test = train_test_split(X2, y2, test_size = 0.20, random_state = 0)
```

```python
model1_1 = MultinomialNB().fit(X1_train,Y1_train)
Y1_pred=model1_1.predict(X1_test)
conf_matr = confusion_matrix(Y1_test,Y1_pred)
accuracy = accuracy_score(Y1_test,Y1_pred)
print(conf_matr)
print(accuracy)
```

```python
model1_2 = MultinomialNB().fit(X2_train,Y2_train)
Y2_pred=model1_2.predict(X2_test)
conf_matr = confusion_matrix(Y2_test,Y2_pred)
accuracy = accuracy_score(Y2_test,Y2_pred)
print(conf_matr)
print(accuracy)
```

```python
model2_1 = DecisionTreeClassifier().fit(X1_train,Y1_train)
```

```python
Y1_pred=model2_1.predict(X1_test)
conf_matr = confusion_matrix(Y1_test,Y1_pred)
accuracy = accuracy_score(Y1_test,Y1_pred)
print(conf_matr)
print(accuracy)
```

```python
model2_2 = DecisionTreeClassifier().fit(X2_train,Y2_train)
```

```python
Y2_pred=model2_2.predict(X2_test)
conf_matr = confusion_matrix(Y2_test,Y2_pred)
accuracy = accuracy_score(Y2_test,Y2_pred)
print(conf_matr)
print(accuracy)
```

```python
model3_1 = RandomForestClassifier().fit(X1_train,Y1_train)
```

```python
Y1_pred=model3_1.predict(X1_test)
conf_matr = confusion_matrix(Y1_test,Y1_pred)
accuracy = accuracy_score(Y1_test,Y1_pred)
print(conf_matr)
print(accuracy)
```
